Now that we have practiced, let's solve this murder!

All we know at the start of our investigation is the <b>murder</b> occurred on <b>Jan.15, 2018</b> and that it took place in <b>SQL City<b>.

```python
#Write a query in SQL to view the crime scene report 
#for the murder occurred on Jan.15, 2018 and that it took place in SQL City.
#Select ALL records FROM the table WHERE...
#this allows me to see the entire description 
pd.set_option('display.max_colwidth', None)

with engine.connect() as connection:  # Establish a connection
    query_murder = text("""SELECT * 
                    FROM crime_scene_report 
                    WHERE type='murder'
                    AND date == 20180115
                    AND city='SQL City'
                    """) #defining the query, from the crime_scene_report, I looked for where type is specifically murder, the date is jan 15, 2018, and the city is SQL City.
    murder = pd.read_sql(query_murder, connection) #Use pandas to read the sql query with the connection to the database



murder
```

Another clue! It looks like we have two witnesses! Now who are they and how do we find them?

```python
#Write two queries in SQL to find each witness in the person table. 
#For the first query as first_witness, select all the records from person where the address_street_name is Northwestern Drive and order by adress_number
#Fetch the first_witness
with engine.connect() as connection:  # Establish a connection
    query_w1 = text("""SELECT * 
                       FROM person
                       WHERE address_street_name = 'Northwestern Dr'
                       ORDER BY address_number DESC
                       LIMIT 1
                       """) #DESC orders address number in decending order (since it was the last house)
                            #LIMIT limits the amount of rows
    w1 = pd.read_sql(query_w1, connection)  # Use pandas to execute the query and fetch results


w1
```

```python
print(f'We have identified our first witness as {w1["name"].iloc[0]}. Now lets find the second!')
```

```python
#For the third query as second_witness,
#Select all the records from person where the address_street_name is Frnaklin Ave and the name includes Annabel. 
#Fetch the second_witness
with engine.connect() as connection:  # Establish a connection
    query_w2 = text("""SELECT * 
                       FROM person
                       WHERE address_street_name = 'Franklin Ave'
                       AND name LIKE '%Annabel%' 
                       """)  #LIKE finds text with the characters we want in it
    w2 = pd.read_sql(query_w2, connection)  # Use pandas to execute the query and fetch results


w2
```

```python
print(f'We have identified our second witness as {w2["name"].iloc[0]}.') 
print("\n What were their witness statements? What did they see?")
```
