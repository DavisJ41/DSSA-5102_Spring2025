```python
import pandas as pd # Reads, writes, shapes, manipulates data
import numpy as np # Basic stats and numerical operations
import matplotlib.pyplot as plt # Creating basic graphs
```

```python
# Load the locally stored CSV file as a dataframe
youtube = pd.read_csv("most_subscribed_youtube_channels.csv" ,delimiter=",")
#making sure the csv file is read in as a dataframe
youtube = pd.DataFrame(youtube)
```

```python
#this will show all of the column names 
youtube.info()
print("\n I have 7 columns in this dataset. \n I have 1000 rows in this dataset.")
youtube.head()
youtube.tail()
youtube.describe()
```

```python
# What are the unique column names?
unique_columns_youtube = pd.Series(youtube.columns).unique()
print("\n The unique columns in this dataset are: \n" + str(unique_columns_youtube))
# Find and print the total number of columns
total_columns_youtube = pd.Series(youtube.columns).count()
print("\n The total number of columns are: " + str(total_columns_youtube))
# Find and print the total number of rows
total_rows_youtube = len(youtube)
print("\n The total number of rows are: " + str(total_rows_youtube))
```

```python
# Look more closely at the data types
for column, dtype in youtube.dtypes.items():
    print(f"{column}: {dtype}")

print("\n It's saying there are two integer types and five object types.")
print("\n After reviewing the file I noticed the commas in three of the columns are turning \n integers into strings.")
```

```python
#removing commas and converting some numerical columns to integers
youtube['video views'] = youtube['video views'].replace({',': ''}, regex=True).astype(int)
youtube['video count'] = youtube['video count'].replace({',': ''}, regex=True).astype(int)
youtube['subscribers'] = youtube['subscribers'].replace({',': ''}, regex=True).astype(int)


for column, dtype in youtube.dtypes.items():
    print(f"{column}: {dtype}")

youtube.head()

print("\n Now there are five integer types and two objects.")
```

```python
# Describe the numerical columns
print("\n This data set contains a lot of unique columns, these columns are labeled: \n rank, subscribers, video views, video count, and started.")
```

```python
# Explore the values of a category column:
print("\n The unique values in a string column labeled 'category' include: \n" + str(youtube['category'].unique()))
```

```python
#creating a graph to represent some of the data 
plt.scatter(youtube['video views'], youtube['subscribers']) 
plt.xlabel('Video Views') #creating a label for the x-axis
plt.ylabel('Subscribers') #creating a label for the y-axis
plt.title('Subscribers vs Video Views') #title
plt.show() #showing the graph 
```

```python
import pandas as pd # Reads, writes, shapes, manipulates data
import numpy as np # Basic stats and numerical operations
import matplotlib.pyplot as plt # Creating basic graphs
```

```python
# Load the locally stored CSV file as a dataframe
nutrition = pd.read_csv("nutrition.csv" ,delimiter=",")
#making sure the csv file is read in as a dataframe
nutrition = pd.DataFrame(nutrition)
```

```python
#this will show all of the column names 
nutrition.info()
```

```python
#checking to see if there are any NA values
print(nutrition.isna().any().any())  #will return True if any Na's exist
```

```python
#removing the first two columns because they are indexed columns that end up stacking on each other
#because they are combined files
nutrition = nutrition.drop(['Unnamed: 0.1', 'Unnamed: 0'] , axis = 1)
```

```python
# Make column names lowercase
nutrition.columns = nutrition.columns.str.lower()
nutrition.info()
```

```python
#changing all columns to have no spaces 
nutrition.columns = nutrition.columns.str.replace(' ', '_')
nutrition.info()
```

```python
#checking the first few rows to see if I want to change anything else
nutrition.head()
```

```python
#changing the dataset so that all of the food is alphebetical order
nutrition = nutrition.sort_values(by = 'food')
nutrition.head()
```

```python
#saving as a new csv file
nutrition.to_csv('clean_nutrition.csv', index = False)
```
